```python
import pandas as pd
import numpy as np
```

```python
df_8 = pd.read_excel("8А_2023_2024.xlsx", header=8)
df_8.head()
```

```python
def clean_df(df, eight_class = True):
    tech_columns = ['Техно 1 гр', 'Техно 2 гр']
    if eight_class:
        df['Техно'] = df[tech_columns[0]].combine_first(df[tech_columns[1]])
    
    columns_to_drop = [
        *tech_columns,
        'Разг о важном', 
        'Россия -  мои горизонты'
    ]
    df = df.drop(columns=columns_to_drop, errors='ignore')
    
    df = df.drop(df.index[:1])
    df = df[df.iloc[:, 1] != "Итог. оценка"]
    
    df = df.rename(columns={'Unnamed: 1': 'Период'})

    if not eight_class:
        df['труд'] = df['труд'].fillna(method='ffill')
    
    return df.reset_index(drop=True)
    
df_8 = clean_df(df_8)
```

```python
df_8.head(10)
```

```python
df_9 = pd.read_excel("9А_2024_2025.xlsx", header=8)
df_9 = clean_df(df_9, False)
```

```python
df_9
```

```python
df1 = df_8.copy()
df2 = df_9.copy()
```

```python
df2 = df2.rename(columns={
    'Вер и Стат': 'ГД',
    'ОБиЗР': 'ОБЖ',
    'труд': 'Техно'
})

for col in df1.columns:
    if col not in df2.columns:
        df2[col] = None  # или pd.NA

for col in df2.columns:
    if col not in df1.columns:
        df1[col] = None  # или pd.NA

combined_df = pd.concat([df1, df2], ignore_index=True)

common_columns = sorted(set(df1.columns) | set(df2.columns))
combined_df = combined_df[common_columns]

combined_df['Муз.'] = combined_df['Муз.'].fillna('Нет данных')
```

```python
combined_df.head(20)
```

```python
new_order = [
    'Период', 'Алг.', 'Биол.', 'Геогр.', 'Геом.', 'ГД', 
    'Англ.яз', 'Инф.', 'Ист.', 'Лит-ра', 'Муз.', 
    'Обществ.', 'ОБЖ', 'Рус.яз.', 'Смысл. чтен.', 
    'Физика', 'Физ-ра', 'Химия', 'Техно'
]
combined_df = combined_df[new_order]
```

```python
combined_df = combined_df.drop(columns=['Муз.'])
```

```python
combined_df.to_csv("8-9 успеваемость.csv", index=False, encoding="utf-8-sig")
```
