```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('tables/10-11_2023-2025.csv')
df.head(10)
```

```python
df.describe()
```

```python
sns.histplot(df)
```

```python
df_long = df.melt( 
    value_vars=[
        'Алг и матанализ', 'Биол.', 'Вер и Стат', 'Геогр.', 'Геом.', 
        'Англ.яз', 'Инф.', 'Ист.', 'Лит-ра', 'Обществ.', 'ОБиЗР', 
        'Практ. по обществ.', 'Рус.яз.', 
        'Совр. литерат', 'Физика', 'Физ-ра', 'Химия'
    ],
    var_name='Предмет',
    value_name='Оценка'
)
```

```python
plt.figure(figsize=(15, 8))
sns.boxplot(data=df_long, x='Предмет', y='Оценка')
plt.xticks(rotation=90)  # Поворот подписей предметов
plt.title('Распределение оценок по предметам')
```

```python
# Выбираем только числовые столбцы с оценками
subjects = [
    'Алг и матанализ', 'Биол.', 'Вер и Стат', 'Геом.', 
    'Физика', 'Химия', 'Инф.'
]

plt.figure(figsize=(12, 8))
sns.heatmap(
    df[subjects].corr(), 
    annot=True, 
    cmap='coolwarm'
).set(title='Корреляция между техническими предметами')
```

```python
sns.lineplot(
    data=df.groupby('Период')['Химия'].mean().reset_index(), 
    x='Период', 
    y='Химия', 
    marker='o'
).set(title='Динамика оценок по химии')
```
