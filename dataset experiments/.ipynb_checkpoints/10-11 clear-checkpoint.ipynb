```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('10-11_2023-2025.csv')
df.head(10)
```

```python
unness_columns = ['ID', 'Класс', 'Период', 'Физ-ра']

subject_columns = [col for col in df.columns if col not in unness_columns]

df_cleaned = df.dropna(how='all', subset=subject_columns)
df_cleaned = df_cleaned.drop(columns=['Практ. по матем.', 'Фин.гр.'])
df_cleaned = df_cleaned.drop(columns=['ID', 'Класс'])
df_cleaned = df_cleaned.reset_index(drop=True)
```

```python
df_cleaned.to_csv("10-11_2023-2025.csv", index=False, encoding="utf-8-sig")
```
